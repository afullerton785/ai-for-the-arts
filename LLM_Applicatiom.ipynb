# The AI Assisted Museum Tour of Music History

## Why
In a theoretical museum of musical history, my Large Language Model (LLM) would help support viewers as they explored. The model would provide a personal experience for each user; varying by age and adapting for those with differing needs

## How LLM's Work

LLM's are trained on huge amounts of data to help them understand the patterns within human language. They perform various tasks using this pattern data and although they dont explicitly understand the language, they can predict what should be said in context thanks to the data they have been trained on. This is how they respond with fluent answers when summarising text, translating, and generating new content.

## How my LLM would Operate
The model would be trained on a large database of text about every exhibit in the museum, but also about extensive knowledge further than just what is in the museum. This allows the model to provide personalised responses incase questions are asked about more than just what is contained in the exhibits. 

The user will be able to decide what role the model will take. That could be simply to guide the user through the museum, or to act as a companion providing consistant anectodes and engaging in conversation as the user explores.

This model would be extremely flexible with whatever the users preferences are and should have extensive enough knowledge to answer any questions about the museum itself, and any further questions users may have. However, as with any LLM the model may produce slight errors or inaccuracies when discussing with the user.

## Daily Use

1. Visitor enters museum and approaches an exhibit centring on the music scene in Glasgow.
2. Visitor then asks the LLM "What famous musicians have come from Glasgow?"
3. LLM accesses database and responds with a list of the most famous musicians from Glasgow then offers further information if wanted.
4. Visitor asks for more specific information about the acts mentioned.
5. LLM then adjusts to the visitors needs and can discuss the artists with a deeper approach.

#### This is essentially the step by step model of how the model would work step-by-step:
1. User input
2. LLM reads input and determines type of response needed (Is it Q&A, conversational, critical)
3. LLM produced output suited to user input.
4. User again inputs something related, suggesting they want more/less input from the model or for things to be explained better.
5. LLM personalises approach to responses now it understands what user wants.
6. LLM produces output to match user desire.

#### Example:
User: "How does a computer make music?"

LLM: "Computers use software to read digital instructions (MIDI) 
that control virtual instruments or recorded audio..."

User: "I don't understand, explain it like I'm 10."

LLM: "Computers use numbers to make the sound!..."

## Enhancement & Inclusion
#### Enhancement:
This LLM would greatly enhance the experience as it would have banks of knowledge not in the exhibit itself, such as information about cultural impacts and underrepresented areas of music history. The model also enhances the experience through active participation, encouraging users to ask questions and delve deeper into the contents of the museum.
##### Inclusion:
The LLM would be extremely accessible to a huge range of people as it would:
- Speak multiple languages
- Be adjustable for different age ranges
- Provide audio-description for the visually impaired
- Have the ability for neurodivergent adaptation

Therefore, the implementation of this LLM in a history of music museum would help to democratise the experience for people of all ages, languages and abilities.

## Reflection

#### Benefits:
- Democratises music knowledge in the museum
- Allows users to dive deeper into content of the exhibits
- Can be personalised to suit users needs
#### Negatives:
- Could produce inaccuracies when discussing further information than strictly what is in exhibits
- May perpetuate western biases depending on the training given
- Potentially replaces human tour guides, removing someone from their passion and livelihood
- Visitors may become overdependant on the AI, rather than interacting with exhibits

#### Steps to fix these:
- The museum could potentially have the AI as an addition working alongside human guides, so there is always someone verifying information also meaning there is no human jobs at risk
- Ensure the model is trained on extensively diverse data to ensure there is no cultural bias present
- Give clear disclaimers throughout that all information should be double checked and nothing the model says can be considered 100% correct.
